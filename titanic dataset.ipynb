# Q.3.  Univariate Analysis - Plot the univariate graphs and find the outliers.Explain how you have handled the outliers by transformation .

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='Survived').map(plt.scatter,'Pclass','Fare').add_legend()
plt.show()
```

# there is only one outlier which lie in first class whose fare cost is above 500.
# second class persons fare are less 100.


```python
ax = sns.violinplot(x="Pclass", y="Fare", data=data)
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue='Survived',size = 5).map(plt.scatter,'Siblings/Spouses Aboard','Pclass').add_legend()
plt.show()
```

```python
sns.distplot(data['Pclass'])
plt.show()
sns.FacetGrid(df, hue = 'Survived', size = 6).map(sns.distplot, 'Pclass').add_legend();
plt.show()
```

```python
sns.boxplot(x='Pclass' , y='Fare', data=data)
plt.show()
```

# Q.5. Multivariate Analysis - Plot the multivariate graphs and find the relationship between the variables. Explain your findings and inference of the graphs. 


```python
sns.pairplot(df,hue='Survived',height=2).add_legend()
plt.show()
```

```python
sns.jointplot(x='Pclass', y='Age',data=df)
plt.show()
```

```python
sns.stripplot(x='Survived',y='Age',data=data,jitter=True);
plt.show()
```

```python
sns.barplot(x='Pclass',y='Fare',data=data);
plt.show()
```

```python
sns.swarmplot(x='Survived',y='Age',data=data);
plt.show()
```
